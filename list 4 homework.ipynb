```python
import numpy as np
import pandas as pd
import io
import statsmodels.formula.api as smf
import scipy.stats as stats
```

```python
data_string = '''Names        Estimate   Std.Error
(Intercept) 3446.830     81.840
totwrk        -0.169      0.018
age            2.714      1.472
south        102.272     41.925
male          87.108     35.173
smsa         -54.188     33.193
yngkid       -13.051     50.459
marr          31.360     42.263
union         11.866     38.186'''
data_string
```

```python
df = pd.read_csv(io.StringIO(data_string), sep ='\s+')
df
```

```python
df['t_nabl'] = np.abs(df['Estimate']/df['Std.Error'])
df
```

```python
alpha = 0.05
t_crit = stats.t.ppf(1-0.05/2, 707 - (8+1)).round(3)
t_crit
```

```python
df['Znachimosty'] = df.apply(lambda x: 'znachim' if x['t_nabl']>t_crit else 'ne_znachim', axis = 1)
df
```

## 2.2 Задача (output equation 2, задача 2.1 в классной работе)

```python
df_lab = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/Labour.csv')
df_lab
```

```python
models1 = smf.ols(data = df_lab, formula = 'np.log(output)~np.log(capital)+np.log(labour)').fit()
models1.summary()
```

```python
models1.summary(alpha = 0.1).tables[1]
```

```python
models1.t_test('np.log(labour)=0.7')
```

```python
beta_cap = 0.7148
ste_cap = 0.023
t_nabl = np.abs((beta_cap-0.7)/ste_cap).round(2)
t_nabl
```

$$
t_{nabl} =0.64
$$

```python
t_crit = stats.t.ppf(1-0.1/2, len(df_lab)-3).round(2)
t_crit
```

$$
t_{crit}=1.65
$$

### ГИПОТЕЗА

$$
H_0:\beta_{labour}=0,5\\
H_1:\beta_{labour}\neq0,5
$$

### ВЫВОД:

## beta_{labour} значительно отличается от 0,7
